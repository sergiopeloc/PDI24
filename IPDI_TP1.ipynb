```python
#importamos libreria imageio para manipular imagenes
import imageio
#importamos para manipular las imagenes
import numpy as np
#importamos para mostrar las imagenes
import matplotlib.pyplot as plt
```

```python
#funcion para pasar de RGB a YIQ
def RGB2YIQ(img):
  yiq=np.zeros(img.shape) #creo una matriz de zeros, del mismo tamaño que la original (img.shape)
  yiq[:,:,0]=np.clip(0.229*img[:,:,0]+0.587*img[:,:,1]+0.114*img[:,:,2],0,1) #banda de la "y" de luminancia // np.clip(operacion,desde,hasta)
  yiq[:,:,1]=np.clip(0.59*img[:,:,0]+-0.27*img[:,:,1]+-0.32*img[:,:,2],-0.5957,0.5957)
  yiq[:,:,2]=np.clip(0.21*img[:,:,0]+-0.52*img[:,:,1]+0.31*img[:,:,2],-0.5226,0.5226)
  return yiq
```

```python
#abrir imagen
#img=imageio.imread('lena512.png')/255 #abro imagen y la normalizo
#img=imageio.imread('image1.png')/255 #abro imagen y la normalizo
img=np.clip(imageio.imread('image1.png')/255.,0,1) #abro imagen y la normalizo
print(img.shape) #saber el tamaño de la imagen (la forma)
print(img.dtype) #saber que tipo de dato

plt.figure(0)
plt.imshow(img) #mostrar la imagen (se muestran los 3 canales)
plt.figure(1)
plt.imshow(img[:,:,0])
plt.figure(2)
plt.imshow(img[:,:,1])
plt.figure(3)
plt.imshow(img[:,:,2])
```

```python
#llamo a la funcion que convierte de RGB a YIQ
yiq=RGB2YIQ(img)

plt.figure(0)
plt.imshow(yiq)
plt.figure(1)
plt.imshow(yiq[:,:,0],'gray')
plt.figure(2)
plt.imshow(yiq[:,:,1],'gray')
plt.figure(3)
plt.imshow(yiq[:,:,2],'gray')

#rgb2=YIQ2RGB(yiq2)
```
