```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../IMAGENES/NUMEROS.bmp',0)


#Aqui se define el tamaño del kernet
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(img,kernel,iterations = 1)
dilation = cv2.dilate(img,kernel,iterations = 1)


(N,M) = img.shape

img_noise = img.copy()

for i in range(N):
  for j in range(M):
    if np.random.rand()>0.99:
      img_noise[i,j] = 255



#Mostrar Imagenes
plt.figure(figsize=(10,5))

#Imagen Original
plt.subplot(1,4,1)
plt.imshow(img,cmap='gray')
plt.title('original')
plt.axis('off')

#Imagen Erocionada
plt.subplot(1 ,4 ,2)
plt.imshow(erosion,cmap='gray')
plt.title('erosion 3x3')
plt.axis('off')

#Imagen Dilatada
plt.subplot(1,4,3)
plt.imshow(dilation,cmap='gray')
plt.title('dilation')
plt.axis('off')

#Imagen 
plt.subplot(1,4,4)
plt.imshow(img_noise,cmap='gray')
plt.title('original + noise (white)')
plt.axis('off')

plt.tight_layout()
plt.show()


```
